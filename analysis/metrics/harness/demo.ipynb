# ``lm-eval`` Demo

Following is an example command to evaluate [LLM360/CrystalChat](https://huggingface.co/LLM360/CrystalChat) on [MMLU](https://arxiv.org/abs/2009.03300) with ``lm-eval``: 

```python
!lm-eval --model hf \
    --model_args pretrained=LLM360/CrystalChat,trust_remote_code=True,dtype=float16 \
    --tasks mmlu \
    --batch_size=16
```
